1: Social Media: Post Engagement Analyzer

```python
def analyze_post_engagement(likes_list):
    total_likes = 0

    # Use a loop for calculations
    for likes in likes_list:
        total_likes += likes

    # Logic to determine status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# Example Usage
post_likes = [400, 350, 300]
analyze_post_engagement(post_likes)
```

3: Agriculture – Rainfall Adequacy Checker

```python
def check_rainfall_adequacy(rainfall_data, required_level):
    total_rainfall = 0

    # Loop to sum the data
    for amount in rainfall_data:
        total_rainfall += amount

    # Calculate average
    average_rainfall = total_rainfall / len(rainfall_data)

    # Conditional logic for status
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print(f"Average Rainfall: {average_rainfall:.0f}")
    print(f"Rainfall Status: {status}")

# Example Usage (Required level: 70)
seasonal_data = [65, 80, 75, 68]
check_rainfall_adequacy(seasonal_data, 70)
```

4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):
    # Sets only store unique values
    unique_usernames = set(usernames)

    # If the set is smaller than the list, a duplicate existed
    if len(unique_usernames) < len(usernames):
        result = "Yes"
    else:
        result = "No"

    print(f"Duplicate Accounts Found: {result}")

# Example Usage
user_list = ["user123", "tech_wizard", "user123", "nature_fan"]
detect_duplicate_accounts(user_list)
```

6: Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(price_list):
    # Logic to filter only prices above ₹2000
    premium_crops = [price for price in price_list if price > 2000]

    # Display the result in the requested format
    print(f"Premium Crops: {premium_crops}")

# Example Usage
all_prices = [1500, 2500, 1800, 3200, 900]
filter_premium_crops(all_prices)
```

7: System Monitoring – Application Health Checker

```python
def evaluate_system_health(error_count):
    # Multi-level conditional statements
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

# Example Usage
evaluate_system_health(7)
```

8: Banking – Daily Transaction Limit Checker

```python
def validate_transaction(amount):
    DAILY_LIMIT = 50000

    # Conditional logic to check against the limit
    if amount <= DAILY_LIMIT:
        status = "Approved"
    else:
        status = "Rejected"

    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

# Example Usage
validate_transaction(60000)
```

9: E-Learning – Student Attendance Eligibility System

```python
def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    days_present = 0

    # Loop to count presence (1 = Present, 0 = Absent)
    for day in attendance_list:
        if day == 1:
            days_present += 1

    attendance_percentage = (days_present / total_days) * 100

    # Eligibility check
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print(f"Attendance Percentage: {attendance_percentage:.1f}")
    print(f"Exam Eligibility: {status}")

# Example Usage (8 present out of 10 days)
attendance_data = [1, 1, 1, 0, 1, 1, 1, 0, 1, 1]
check_exam_eligibility(attendance_data)
```

10: Smart Electricity Bill Analyzer

```python
def analyze_electricity_bill(units):
    bill = 0

    # Billing Slabs Logic
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Usage Classification Logic
    if bill < 500:
        classification = "Low Usage"
    elif 500 <= bill <= 1500:
        classification = "Moderate Usage"
    else:
        classification = "High Usage"

    return bill, classification

# Example Usage
units_consumed = 250
total_bill, usage_status = analyze_electricity_bill(units_consumed)

print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Level: {usage_status}")
```
