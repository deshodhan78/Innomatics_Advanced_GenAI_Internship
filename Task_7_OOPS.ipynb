**1: Smart Light Controller**  
Create a class to control a smart light that can be turned ON or OFF and display its status.  
Real-Time Use:
Smart home automation systems  
Sample Input:  
Light Name: Bedroom Light  
Action: ON  
Expected Output:  
Bedroom Light is ON

```python
class Light:
  def __init__(self, name, status):
    self.name=name
    self.status=status

  def show_status(self):
      print(self.name,' is ',self.status)

light=Light('Bedroom Light','OFF')
light.show_status()

light=Light('Bedroom Light','ON')
light.show_status()
```

2: Employee ID Card System:  
Create a class that stores employee name, ID, and department and displays ID card details.  
Real-Time Use:  
Corporate HR systems  

**Sample Input**  
Employee Name: Rahul  
Employee ID: EMP102  
Department: IT  

**Expected Output**  
Employee ID Card  
Name: Rahul  
ID: EMP102  
Department: IT


```python
from os import name
class Employee:

  def __init__(self, id, name, dep):
    self.ID=id
    self.name=name
    self.dep=dep

  def show_details(self):
    print("Employee ID Card")
    print(self.ID, self.name, self.dep)

emp1=Employee('EMP102','Deepak','GenAI')
emp1.show_details()

emp2=Employee('EMP103','Santosh','IT')
emp2.show_details()
```

 **3: Mobile Contact Record**  
Create a class to store contact name and phone number and display contact information.  

**Real-Time Use**  
Mobile contact applications  

**Sample Input**  
Contact Name: Anita  
Phone Number: 9876543210  

**Expected Output**  
Contact Saved  
Name: Anita  
Phone: 9876543210

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

contact=Contact("Deepak", 7894561235)
contact.display_contact()
```

**4: Product Price Tag Generator**  
Create a class that stores product name and price and prints a formatted price tag.  

**Real-Time Use**  
Retail billing systems

**Sample Input**  
Product Name: Headphones  
Price: 2499  
**Expected Output**  
Product: Headphones  
Price: ₹2499

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

product=Product("HEadphones",500)
product.print_tag()
```

**5: Movie Rating Display System**  
Create a class that stores movie name and rating and displays movie details.  
**Real-Time Use**  
Streaming platforms  
**Sample Input**  
Movie Name: Inception  
Rating: 4.8  
**Expected Output**  
Movie: Inception  
Rating: 4.8 / 5

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def show_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

movie=Movie('Inception',4.8)
movie.show_details()
movie=Movie('Dark knight',4.7)
movie.show_details()
```

**6: Delivery Address Manager**
Create a class that stores customer name and delivery address and prints delivery details.  

**Real-Time Use**  
Courier and logistics platforms  
**Sample Input**  
Customer Name: Suman  
Address: Hyderabad  
**Expected Output**  
Delivery Details  
Customer: Suman  
Address: Hyderabad

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def print_details(self):
        print("Delivery Details")
        print('-----------')
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")

deli=Delivery('Deepak','Hyderabad')
deli.print_details()
```
