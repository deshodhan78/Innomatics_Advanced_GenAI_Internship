```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from typing import List

```

TASK 1: Replace Hardcoded Prompts

```python
def explain_topic(topic: str) -> str:
    prompt_template = PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} in simple terms for beginners"
    )
    return prompt_template.format(topic=topic)

print("→", explain_topic("Quantum Computing"))
print("→", explain_topic("Blockchain"))
```

TASK 2: Multi-Input Prompt System

```python
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

for i, case in enumerate(test_cases, 1):
    prompt = multi_input_template.format(**case)
    print(f"Test Case {i}: {prompt}")
```

TASK 3: Prompt Variations Engine

```python
topic = "Machine Learning"

teaching_template = PromptTemplate.from_template("Explain {topic} clearly step by step")
interview_template = PromptTemplate.from_template("Ask 3 questions about {topic}")
storytelling_template = PromptTemplate.from_template("Explain {topic} as a story")

print("Teaching  →", teaching_template.format(topic=topic))
print("Interview →", interview_template.format(topic=topic))
print("Story     →", storytelling_template.format(topic=topic))
```

TASK 5: Input Validation

```python
def validate_inputs(audience: str, tone: str):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    if audience not in valid_audiences:
        print(f"Invalid audience '{audience}'. Defaulting to 'beginner'.")
        audience = "beginner"
    if tone not in valid_tones:
        print(f"Invalid tone '{tone}'. Defaulting to 'casual'.")
        tone = "casual"

    return audience, tone

validate_inputs("Learner", "formal")
```

TASK 4: ChatPromptTemplate System

```python
def create_chat_prompt(role: str, topic: str):
    role_behaviors = {
        "teacher": "You are a patient and clear teacher who explains concepts simply.",
        "interviewer": "You are a technical interviewer evaluating deep understanding.",
        "motivator": "You are an encouraging coach who motivates while teaching."
    }

    system_msg = role_behaviors.get(role.lower(), role_behaviors["teacher"])

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("human", "Help me understand {topic}.")
    ])

    return chat_template.format_messages(topic=topic)

messages = create_chat_prompt("teacher", "Neural Networks")
for msg in messages:
    print(f"{msg.type.upper()}: {msg.content}")
```

TASK 6: Prompt Generator App

```python
def generate_prompt(topic: str, audience: str = "beginner",
                   tone: str = "casual", style: str = "teaching") -> str:

    audience, tone = validate_inputs(audience, tone)

    style_templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} tone. Do it clearly step by step.",
        "interview": "Act as an interviewer. Ask 3 questions about {topic} for {audience} in a {tone} tone.",
        "storytelling": "Explain {topic} as an engaging story for {audience} in a {tone} tone."
    }

    if style not in style_templates:
        raise ValueError("Style must be 'teaching', 'interview', or 'storytelling'")

    template = PromptTemplate.from_template(style_templates[style])
    return template.format(topic=topic, audience=audience, tone=tone)

print("Example:", generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

TASK 7: Reusability Test

```python
reusable_template = PromptTemplate.from_template(
    "Explain {topic} for {audience} in a {tone} tone"
)

test_inputs = [
    {"topic": "Quantum Computing", "audience": "beginner", "tone": "fun"},
    {"topic": "Blockchain", "audience": "expert", "tone": "formal"},
    {"topic": "Python", "audience": "kids", "tone": "casual"},
    {"topic": "Data Science", "audience": "intermediate", "tone": "fun"},
    {"topic": "Cyber Security", "audience": "beginner", "tone": "formal"}
]

for i, inp in enumerate(test_inputs, 1):
    print(f"{i}. {reusable_template.format(**inp)}")
```
