1.User Login Check

Problem Statement:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"

Requirements:
Print "Login Successful" if both username and password match
Otherwise print "Invalid Credentials"

```python
corr_username="admin"
corr_password="1234"

u_input = input("Enter Username: ")
p_input = input("Enter Password: ")

if(u_input==corr_username and p_input==corr_password):
  print("Login Successful")
else:
    print("Invalid Credentials")
```

2.Pass / Fail Analyzer

Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:
A student passes if marks â‰¥ 50

Count the total number of pass students

Count the total number of fail students

Print the final result clearly


```python
# List of student marks
marks = [45, 78, 90, 33, 60]

# Let's initialize counters
pass_count = 0
fail_count = 0

# Passing threshold
passing_score = 50

# Iterate through the list
for score in marks:
    if score >= passing_score:
        pass_count += 1
    else:
        fail_count += 1

print("-" * 25)
print(f"Total Students: {len(marks)}")
print(f"Passed: {pass_count}")
print(f"Failed: {fail_count}")
print("-" * 25)
```

3.Simple Data Cleaner

Problem Statement:

Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

Requirements:

Remove extra spaces

Convert all names to lowercase

Print the cleaned list

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []

for name in names:
    # Clean whitespace with strip() and convert to lowercase with lower()
    clean_name = name.strip().lower()
    cleaned_names.append(clean_name)

print("Cleaned List:", cleaned_names)
```

4.Message Length Analyzer

Problem Statement:

Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:

Print the length of each message

Flag messages longer than 10 characters


```python
messages = ["Hi", "Welcome to the platform", "OK"]

# Threshold for flagging
limit = 10

print("Message Analysis:")
print("-" * 30)

for msg in messages:
    msg_length = len(msg)

    # Check if the message exceeds the limit
    if msg_length > limit:
        status = "[LONG MESSAGE]"
    else:
        status = "[SHORT MESSAGE]"

    print(f"Length: {msg_length:2} | {status} | '{msg}'")

print("-" * 30)
```

5.Error Message Detector

Problem Statement:

Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:

Count the number of "ERROR" entries

Print the total error count


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print(f"Total Error Count: {error_count}")
```
